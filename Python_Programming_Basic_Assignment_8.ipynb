# Q1.
Write a Python Program to Add Two Matrices?

```python
Matrix1=[[1,2],[3,4]]
Matrix2=[[2,3],[3,5]]
Matrix3=[[0,0],[0,0]]
for i in range(0,len(Matrix1)):
    for j in range(0,len(Matrix2)):
        Matrix3[i][j]=Matrix1[i][j]+Matrix2[i][j]
print("The sum of matrices is: ",Matrix3)
```

# Q2.
Write a Python Program to Multiply Two Matrices?

```python
import numpy as np
Matrix1=np.array([[1,2],[3,4]])
Matrix2=np.array([[2,3],[3,5]])
print("The Vector Multiplication two of Matrices is: ",Matrix1.dot(Matrix2))
print("The Elementwise Multiplication two of Matrices is: ",Matrix1*Matrix2)
```

# Q3.
Write a Python Program to Transpose a Matrix?

```python
import numpy as np
Matrix1=np.array([[1,2],[3,4]])
print("Before Matrix Transpose: ",Matrix1)
print("After Matrix Transpose: ",Matrix1.transpose())
```

# Q4.
Write a Python Program to Sort Words in Alphabetic Order?

```python
length=int(input("Enter Length of Array: "))
Alphabets=[]
for i in range(0,length):
    words=str(input("Enter words: "))
    Alphabets.append(words)
Alphabets.sort()
print("Array After Sorting in Alphabetical Order: ",*Alphabets)
```

# Q5.
Write a Python Program to Remove Punctuation From a String?

```python
import re #regex
String="HI @Deepu, I want you to remove the$se E*xtr@a Pun#ctu@ation$s."
String2=re.sub(r'[^\w\s]','',String)
print("String after removing Punctuation: ",String2)
```
