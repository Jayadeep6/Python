# Question 1:
Please write a program using generator to print the numbers which can be divisible by 5 and 7 between 0 and n in comma separated form while n is input by console.
Example:
If the following n is given as input to the program:
100
Then, the output of the program should be:
0,35,70

```python
def divisible(number):
    for i in range(0,number+1):
        if i%5==0 and i%7==0:
            yield i
number=int(input("Enter then value: "))
array=[]
for i in divisible(number):
    array.append(str(i))
print(",".join(array))
```

# Question 2:
Please write a program using generator to print the even numbers between 0 and n in comma separated form while n is input by console.
Example:
If the following n is given as input to the program:
10
Then, the output of the program should be:
0,2,4,6,8,10

```python
def even(number):
    for i in range(number+1):
        if i%2==0:
            yield i
number=int(input("Enter the number: "))
array=[]
for i in even(number):
    array.append(str(i))
print(",".join(array))
```

# Question 3:
The Fibonacci Sequence is computed based on the following formula:

f(n)=0 if n=0

f(n)=1 if n=1

f(n)=f(n-1)+f(n-2) if n>1

Please write a program using list comprehension to print the Fibonacci Sequence in comma separated form with a given n input by console.

Example:
If the following n is given as input to the program:

7

Then, the output of the program should be:
0,1,1,2,3,5,8,13

```python
def fibonacci(number):
    f0,f1=0,1
    if number==1:
        yield f0
    else:
        for i in range(number+1):
            yield f0
            f0,f1=f1,f0+f1
number=int(input("Enter a number: "))
lst=[]
if number>1:
    for i in fibonacci(number):
        lst.append(str(i))
    print(",".join(lst))
else:
    print("Please enter a positive number")
```

# Question 4:

Example:
If the following email address is given as input to the program:

john@google.com

Then, the output of the program should be:

john

```python
def username(email):
    user=[]
    company=[]
    user=email.split("@")
    company=user[1].split(".")
    if user[0].isalpha() and company[0].isalpha():
        return user[0]
    else:
        return "Email can only contain letters"
email=str(input("Enter email id: "))
print(username(email))
```
