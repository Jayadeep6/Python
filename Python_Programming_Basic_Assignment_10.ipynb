# Q1.
Write a Python program to find sum of elements in list?

```python
length=int(input("Enter the length: "))
lists=[]
sums=0
for i in range(length):
    value=int(input("Enter the value: "))
    lists.append(value)
for i in lists:
    sums+= i
print("The sum of elements in lits is: ",sums)
```

# Q6.
Write a Python program to find N largest elements from a list?

```python
length=int(input("Enter length: "))
lists=[]
for i in range(0,length):
    value=int(input("Enter value: "))
    lists.append(value)
print("List: ",lists)
lists.sort()
lists.reverse()
n=int(input("Enter n largest elements value: "))
print("The",n,"largest elements from a list are: ",lists[0:n])
```

## Q7.
Write a Python program to print even numbers in a list?

```python
length=int(input("Enter length: "))
lists=[]
for i in range(0,length):
    value=int(input("Enter value: "))
    lists.append(value)
print("The even numbers in the given list: ")
for i in range(length):
    if(lists[i]%2==0):
        print(lists[i])
```

# Q8.

# Q10.
Write a Python program to Cloning or Copying a list?

```python
length=int(input("Enter length: "))
List=[]
Copy_List=[]
for i in range(0,length):
    value=int(input("Enter value: "))
    List.append(value)
print("The Original list: ",List)
for i in List:
    Copy_List.append(i)
print("The New List after cloning or copying original List: ",Copy_List)
```

# Q11.
Write a Python program to Count occurrences of an element in a list?

```python
length=int(input("Enter length: "))
List=[]
count=0
for i in range(0,length):
    value=int(input("Enter value: "))
    List.append(value)
print("The Original list: ",List)
element=int(input("Enter the element to check its occurences: "))
for i in List:
    if i==element:
        count+=1
print("The Occurances of",element,"in list is: ",count)
```
