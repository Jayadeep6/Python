# Q1.	
Write a Python Program to find sum of array?


```python
length= int(input("Enter the length of array: "))
array=[]
for i in range(length):
    value=int(input("Enter value of array: "))
    array.append(value)
sum_of_Array=0
for i in array:
    sum_of_Array+=i
print("Sum of Array is:",sum_of_Array)
```

# Q2.
Write a Python Program to find largest element in an array?

```python
length= int(input("Enter the length of array: "))
array=[]
for i in range(length):
    value=int(input("Enter value of array: "))
    array.append(value)
print("The largest element of array is: ",max(array))
```

# Q3.
Write a Python Program for array rotation?

```python
length= int(input("Enter the length of array: "))
array=[]
for i in range(length):
    value=int(input("Enter value of array: "))
    array.append(value)
print('Actual Array: ',array)
rotation=int(input("Enter the value for array rotation: "))
if (rotation<length):
    print("Array after Rotation: ",array[rotation:]+array[:rotation])
else:
    print("please enter the number less than length of array")
```

# Q4.	
Write a Python Program to Split the array and add the first part to the end?

```python
length=int(input("Enter the length of array: "))
array=[]
for i in range(0,length):
    value=int(input('Enter Value of array: '))
    array.append(value)
print("Actual Array is: ",array)
first_half=length//2
print("After Splitting the array: ",array[first_half:]+array[:first_half])
```

# Q5.
Write a Python Program to check if given array is Monotonic?

```python
length=int(input("Enter the length of Array: "))
array=[]
count_increase=0
count_decrease=0
for i in range(0,length):
    value=int(input("Enter the Array Value: "))
    array.append(value)
if all(array[i-1]<array[i] for i in range(1,length)):
    print("Array is Monotonic Increasing")
elif all(array[i-1]>array[i] for i in range(1,length)):
    print("Array is Monotonic Decreasing")
else:
    print("Array is not Monotonic")
```
