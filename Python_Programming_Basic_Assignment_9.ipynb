# Q1.
Write a Python program to check if the given number is a Disarium Number?

```python
number=int(input("Enter the number: "))
string=str(number)
sums=0
for i in range(0,len(string)):
    sums+=int(string[i])**(i+1)
if(sums==number):
    print("The given number",number,"is a Disarium Number")
else:
    print("The given number",number,"is NOT a Disarium Number")
```

# Q2.
Write a Python program to print all disarium numbers between 1 to 100?

```python
sums=0
i=1
print("The Disarium numbers between 1 and 100 are: ")
while(i<=100):
    string=str(i)
    for j in range(0,len(string)):
        sums+=int(string[j])**(j+1)
        j+=1
    if(sums==i):
        print(i)
    sums=0
    i+=1
```

# Q3.
Write a Python program to check if the given number is Happy Number?

```python
def happy(string):
    sums=0
    for i in range(0,len(string)):
        sums+=int(string[i])**2
    if(sums >9):
        happy(str(sums))
    else:
        if(sums==1):
            print("Given number is Happy Number")
        else:
            print("Given number is NOT a Happy Number")
number=int(input("Enter a number: "))
string=str(number) 
happy(string)
```

# Q4.
Write a Python program to print all happy numbers between 1 and 100? 

```python
def isHappyNumber(num):    
    rem = sum = 0;      
    while(num > 0):    
        rem = num%10;    
        sum = sum + (rem*rem);    
        num = num//10;    
    return sum;    
print("List of happy numbers between 1 and 100: ");    
for i in range(1, 101):    
    result = i;    
    while(result != 1 and result != 4):    
        result = isHappyNumber(result);       
    if(result == 1):    
        print(i)
```

# Q5.
Write a Python program to determine whether the given number is a Harshad Number?

```python
number=int(input("Enter number: "))
string=str(number)
sums=0
for i in range(0,len(string)):
    sums +=int(string[i])
if number%sums==0:
    print("Given number is a Harshad Number")
else:
    print("Given number is NOT a Harshad Number")
```

# Q6.
Write a Python program to print all pronic numbers between 1 and 100?

```python
def pronic(number):
    for j in range(1,number//2):
        if(number == j*(j+1)):
            print(number)
print("The pronic numbers between 1 to 100: ")
for i in range(1,101):
    pronic(i)
```
