# Q1.	
Write a Python Program to Display Fibonacci Sequence Using Recursion?

```python
def fibonacci(nterm):
    if(nterm<=1):
        return nterm
    else:
        return(fibonacci(nterm-1)+fibonacci(nterm -2))
nterm=int(input("Enter the nterm of series: "))
if(nterm<=0):
    print("Please enter a Positive Number")
else:
    print("Fibonacci Series is: ")
    for i in range(nterm):
        print(fibonacci(i))
```

# Q2
Write a Python Program to Find Factorial of Number Using Recursion?

```python
def factorial(number):
    factorial=1
    for i in range(1,number+1):
        factorial *= i
    return factorial
number=int(input("Enter a number: "))
print(factorial(number))
```

# Q3
Write a Python Program to calculate your Body Mass Index?

```python
weight=float(input("Enter weigth in kgs: "))
height=float(input("Enter height in centimeters: "))
meters= height/100
Bmi=weight/(meters*meters)
print("Your Body Mass Index is: ",round(Bmi,2))
```

# Q5.	
Write a Python Program for cube sum of first n natural numbers?

```python
n=int(input("Enter n value: "))
sum=1
for i in range(2,n+1):
    sum+=(i*i*i)
print("Cube Sum of first",n,"natural numbers is:",sum)
```
