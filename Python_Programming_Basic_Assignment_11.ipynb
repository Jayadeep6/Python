# Q1.
Write a Python program to find words which are greater than given length k?

```python
length=int(input("Enter the number of words: "))
List=[]
print("Enter words: ")
for i in range(0,length):
    word=str(input())
    List.append(word)
k=int(input("Enter the k value to find words which are greater than given length: "))
for i in List:
    if(len(i)>k):
        print(i)
```

# Q4.
Write a Python to check if a given string is binary string or not?

```python
string=str(input("Enter string: "))
binary="01"
count=0
for i in string:
    if i not in binary:
        count+=1
        break
if count:
    print(string,"is a NOT binary string")
else:
    print(string,"is a binary string")
```

# Q5.
Write a Python program to find uncommon words from two Strings?

```python
string1=str(input("Enter string1: "))
string2=str(input("Enter string2: "))
string1=string1.split()
string2=string2.split()
print("The Uncommon words from two strings are: ")
for i in string1:
    if i not in string2:
        print(i)
for i in string2:
    if i not in string1:
        print(i)
```

# Q6.
Write a Python to find all duplicate characters in string?

```python
import re
string=str(input("Enter string: "))
print("The duplicate characters in string: ")
for i in range(0,len(string)):
    for j in range(i+1,len(string)):
        f= re.findall(string[j],string[i])
        for k in f:
            print(k)
```

# Q7.
Write a Python Program to check if a string contains any special character?

```python
import re
string=str(input("Enter string: "))
if(bool(re.match('^[a-zA-Z0-9]*$',string))==True):
    print("String does not have any special characters")
else:
    print("String have special charcters")
```
