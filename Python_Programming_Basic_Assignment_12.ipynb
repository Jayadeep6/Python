# Q1.
Write a Python program to Extract Unique values dictionary values?

```python
List=[]
dictionary={1:'p',2:'r',3:'o',4:'g',5:'r',6:'a',7:'m',8:'m',9:'i',10:'n',11:'g'}
for i in dictionary.values():
    if i not in List:
        List.append(i)
        print(i)
```

# Q2.	
Write a Python program to find the sum of all items in a dictionary?

```python
dictionary={1:10,2:20,3:30,4:40,5:50}
sums=0
for i in dictionary.values():
    sums+=i
print("The sum of all items in given dictionary is: ",sums)
```

# Q6.
Write a Python program to check order of character in string using OrderedDict()?

```python
from collections import OrderedDict 
def check_order(string, pattern): 
    dict = OrderedDict.fromkeys(string) 
    ptrlen = 0
    for key,value in dict.items(): 
        if (key == pattern[ptrlen]): 
            ptrlen = ptrlen + 1
        if (ptrlen == (len(pattern))): 
            return 'true'
    return 'false'
print(check_order("hello python","hp"))
print(check_order("This is jupyter","Tijh"))
```
