# Question 4:
Please write a program to generate all sentences where subject is in ["I", "You"] and verb is in ["Play", "Love"] and the object is in ["Hockey","Football"].


```python
subject=["I","You"]
verb=["Play","Love"]
Object=["Hockey","Football"]
for i in range(0,len(subject)):
    for j in range(0,len(verb)):
        for k in range(0,len(Object)):
            sentence=""
            sentence+=("%s %s %s." % (subject[i],verb[j],Object[k]))
            print(sentence)   
```

# Question 5:
Please write a program to compress and decompress the string "hello world!hello world!hello world!hello world!".

 

```python
def compress(string):
    i = 0
    compressed = ""
    length = len(string)
    while i != length:
        count = 1
        while (i < length-1) and (string[i] == string[i+1]):
            count += 1
            i += 1
        if count == 1:
            compressed += str(string[i])
        else:
            compressed += str(string[i]) + str(count)
        i += 1
    return compressed 
def decompress(compressed):
    decompressed=""
    for i in range(len(compressed)):
        if compressed[i].isnumeric()==False:
            decompressed+=compressed[i]
        else:
            for j in range(int(compressed[i])-1):
                decompressed+=compressed[i-1]
    return decompressed
        
string = "hello world!hello world!hello world!hello world!"
compressed=compress(string)
print(compressed)
print(decompress(compressed))
```

# Question 6:
Please write a binary search function which searches an item in a sorted list. The function should return the index of element to be searched in the list.


```python
def binary(array,item,low,high):
    mid = (low+high)//2
    if low<=high:
        if item == array[mid]:
            return mid
        elif item<array[mid]:
            return binary(array,item,low,mid-1)
        else:
            return binary(array,item,mid+1,high)
    else:
        return -1
       
            
array=[]
length=int(input("Enter the length of array"))
print("Enter values of array: ")
for i in range(length):
    numbers=int(input())
    array.append(numbers)
array.sort()
print(array)
item=int(input("Enter the value of item to search: "))
low =0
high=len(array)-1
index=binary(array,item,low,high)
if index==-1:
    print(item,"is not found in array")
else:
    print(item,"is found at: ",index)
```
