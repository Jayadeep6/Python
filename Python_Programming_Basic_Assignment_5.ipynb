# Q1.
Write a Python Program to Find LCM?

```python
number1=int(input("Enter first number: "))
number2=int(input("Enter second number: "))
greatest=max(number1,number2)
while(True):
    if(greatest%number1==0 and greatest%number2==0):
        print("Lcm of",number1,"and",number2,"is: ",greatest)
        break
    greatest+=1
```

# Q2.	
Write a Python Program to Find HCF?

```python
number1=int(input("Enter first number: "))
number2=int(input("Enter second number: "))
smallest=min(number1,number2)
for i in range(1,smallest+1):
    if(number1%i==0 and number2%i==0):
        hcf=i
print("HCF of",number1,"and",number2,"is: ",hcf)
```

# Q3.
Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?

```python
number=int(input("Enter a Decimal Number: "))
print("Binary Number is: ",bin(number).replace("0b",""))
print("Octal Number is: ",oct(number))
print("Hexadecimal Number is",hex(number))
```

# Q4.
Write a Python Program To Find ASCII value of a character?

```python
character=input("Enter a value: ")
print("ASCII Value of a character is: ",ord(character))
```

# Q5.
Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?

```python
number1=int(input("enter a number : "))
number2=int(input("enter a number : "))
print("Enter any one of the opertor : "
     " + for addition,"
     " - for substraction,"
     " / for Divison,"
     " * for Multiplication")
operator=input("Enter operator u want to perform: ")
if(operator=='+'):
    print("Sum of given numbers",number1,"and",number2,"is:",number1+number2)
elif(operator=='-'):
    print("Substraction of given numbers",number1,"and",number2,"is: ",abs(number1-number2))
elif(operator=='/'):
    print("Division of given numbers",number1,"and",number2,"is:",number1/number2)
elif(operator=='*'):
    print("Multiplication of given numbers",number1,"and",number2,"is:",number1*number2)
else:
    print("Please enter a valid operator")
```
